```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('penguins.csv')
sns.pairplot(data=data, hue='species')
plt.title('Pairplot of Penguin Data')
plt.show()

```

```python
# Boxplot of body mass grouped by species
sns.boxplot(x='species', y='body_mass_g', data=data)
plt.title('Boxplot of Body Mass by Species')
plt.show()
```

```python
species_counts = data.groupby(['species', 'island']).size().unstack()

species_counts.plot(kind='bar', stacked=True)
plt.title('Distribution of Penguin Species Across Islands')
plt.xlabel('Island')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['bill_length_mm'], data['bill_depth_mm'], alpha=0.7)
plt.title('Scatter Plot of Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['body_mass_g'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Body Mass')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
